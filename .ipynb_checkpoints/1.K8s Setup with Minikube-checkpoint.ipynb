## Minikube

____

This is the easiest way to setup k8s on a local system

We have mentioned all the component of K8s, that make up a master and worker node.

So here is the distinction, of how the components are split.

It will setup a lot of time and effort to setup all the components.



Minikube bundles all these components into a single image:

Providing us with a single preconfigured k8s cluster, so we can get going in minutes. 

Everything is packed into an iso image, which is available online for download




But you do not need to download it yourself.
 

Minikube(minikube.exe) provides a exe utility that will automatically download the iso, and deploy it into on virtualbox. So you will need to have a hypervisor installed on your system

To interact with the K8s cluster we need the kubectl command line tool.(must also be installed on machine)

So we need the following for this to work

1. We need a hypervisor (Virtual machine)
2. Minikube.exe
3. Kubectl command line utility




## Minikube utility

____

Here we going to deploy a basic version of K8s using the minikube utility.

We go to: 


1. http://kubernetes.io/
2. Click documentation (we need instructions to install minikube)
3. Click Setup, Getting Started, Learning Environment.
4. Click Installing K8s with Minikube
5. Click Installation

We first need to install Kubectl command line utility

```
curl -LO https://storage.googleapis.com/kubernetes-release/release/`curl -s https://storage.googleapis.com/kubernetes-release/release/stable.txt`/bin/linux/amd64/kubectl

```

Make it executable

```
chmod +x ./kubectl
```

Move it to the correct location

```
sudo mv ./kubectl /usr/local/bin/kubectl

```

And test the version

```
kubectl version --client
```

Download minikube now

```
curl -Lo minikube https://storage.googleapis.com/minikube/releases/latest/minikube-linux-amd64  && chmod +x minikube
```

Then install and add to path minikube

```
sudo mkdir -p /usr/local/bin/
sudo install minikube /usr/local/bin/

```

Download abd created VM image automatically in virtualbox
```
minikube start
```

Output

```
  minikube v1.10.1 on Debian 10.4
  Automatically selected the virtualbox driver
  Downloading VM boot image ...
    > minikube-v1.10.0.iso.sha256: 65 B / 65 B [-------------] 100.00% ? p/s 0s
    > minikube-v1.10.0.iso: 174.99 MiB / 174.99 MiB  100.00% 2.26 MiB p/s 1m17s
  Starting control plane node minikube in cluster minikube
  Downloading Kubernetes v1.18.2 preload ...
    > preloaded-images-k8s-v3-v1.18.2-docker-overlay2-amd64.tar.lz4: 525.43 MiB
  Creating virtualbox VM (CPUs=2, Memory=2200MB, Disk=20000MB) ...
  Preparing Kubernetes v1.18.2 on Docker 19.03.8 ...
  Verifying Kubernetes components...
  Enabled addons: default-storageclass, storage-provisioner
  Done! kubectl is now configured to use "minikube"

```



Go back to Quickstart and issue the following command

Now we can interact with the cluster by using kubectl:


Check the nodes

```
kubectl get nodes
```


Lets create a k8s deployment, by using an existing image named echoserver (simple http server), that is exposed on port 8080

Must first create a deployment and then run afterwards

```
kubectl create deployment hello-minikube --image=k8s.gcr.io/echoserver:1.10 
```

then 
```
kubectl run hello-minikube --image=k8s.gcr.io/echoserver:1.10 --port=8080

```

Now we have pod running


```
kubectl get pods
```

This is web app, so we need to expose eth app on port


```
kubectl expose deployment hello-minikube --type=NodePort --port=8080

```

Then will get result: service/hello-minikube exposed

And how do I access the webpage:

We run the following command, which will give us a webpage

```
minikube service hello-minikube --url
```

This will give the output: http://192.168.99.100:32019/ (this is the IP of the minikube vm)

Then we need to cleanup after us:

```
kubectl delete services hello-minikube (delete minikube service)
kubectl delete deployment hello-minikube (delete minikube deployment)
minikube stop (stop the minikube cluster)
```
