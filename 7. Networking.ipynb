## Setup and more
___

In K8s world an ip is assigned to POD (not a container, as per the docker world), each pod gets its own ip.




And they access the other pods using an internal netork:



We have the following problem when when the nodes are setup individually 




We need the following networking requirements
1. All containers/POD mus be able to communicate with each other, without using NAT.
2. All nodes can communicate with all containers and vice versa without NAT

K8s expect us to setup a network solution with meeting this criteria. But this is quite a bit of work to do. However, there are some help....

Main Point here: We do not have to setup networking ourselves it is done by third party provders, and prebuilt solutions: (we will be using calico)






These providers manages the networks and ips in my nodes.
